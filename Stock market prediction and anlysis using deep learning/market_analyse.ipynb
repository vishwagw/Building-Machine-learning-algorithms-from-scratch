```python
#importing libraires 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#load input dataset 1.
goog_in = pd.read_csv('./Dataset/GOOGL_2006-01-01_to_2018-01-01.csv')
#check the data set by head.
goog_in.head()
```

```python
#input dataset2.
amaz_in = pd.read_csv('./Dataset/AMZN_2006-01-01_to_2018-01-01.csv')
amaz_in.head()
```

```python
#dataset 3.
ibm_in = pd.read_csv('./Dataset/IBM_2006-01-01_to_2018-01-01.csv')
ibm_in.head()
```

```python
#dataset 4
msft_in = pd.read_csv('./Dataset/MSFT_2006-01-01_to_2018-01-01.csv')
msft_in.head()
```

```python

#for first figure:
fig = px.histogram(goog_in, 
                   x='Close', 
                   marginal='box', 
                   nbins=47, 
                   title='Distribution of Close')
fig.update_layout(bargap=0.1)
#showing the output
fig.show()
```

```python
#change the colour to red.
fig = px.histogram(goog_in, 
                   x='Open', 
                   marginal='box', 
                   color_discrete_sequence=['red'], 
                   title='Distribution of open')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.scatter(goog_in, 
                 x='Open', 
                 y='Close', 
                 
                 opacity=0.8, 
                  
                 title='Open vs. Close')
fig.update_traces(marker_size=5)
fig.show()
```

```python
#now the correlation analysis
#lets check the correlation analysis.
goog_in.corr()
```

```python
msft_in['2008':'2018'].plot(subplots=True, figsize=(10,12))
plt.title('Microsoft stock attributes from 2008 to 2018')
plt.show()
```

```python
amaz_in['2008':'2018'].plot(subplots=True, figsize=(10,12))
plt.title('Amazon stock attributes from 2008 to 2018')
plt.show()
```

```python
ibm_in['2008':'2018'].plot(subplots=True, figsize=(10,12))
plt.title('IBM stock attributes from 2008 to 2018')
plt.show()
```

```python
#plotting 
goog_in.High.plot()
amaz_in.High.plot()
msft_in.High.plot()
ibm_in.High.plot()
plt.legend(['Google', 'Microsoft', 'Amazon', 'Ibm'])
plt.show()
```

```python
# Expanding window functions -msft
msft_mean = msft_in.High.expanding().mean()
msft_std = msft_in.High.expanding().std()
msft_in.High.plot()
msft_mean.plot()
msft_std.plot()
plt.legend(['High','Expanding Mean','Expanding Standard Deviation'])
plt.title('Microsoft')
plt.show()
```

```python
#expanding window function - google
google_mean = goog_in.High.expanding().mean()
google_std = goog_in.High.expanding().std()
goog_in.High.plot()
google_mean.plot()
google_std.plot()
plt.legend(['High','Expanding Mean','Expanding Standard Deviation'])
plt.title('Google')
plt.show()
```

```python
# Expanding window functions - amazon
amazon_mean = amaz_in.High.expanding().mean()
amazon_std = amaz_in.High.expanding().std()
amaz_in.High.plot()
amazon_mean.plot()
amazon_std.plot()
plt.legend(['High','Expanding Mean','Expanding Standard Deviation'])
plt.title('Amazon')
plt.show()
```
